```python
 import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# CSV 파일 로드
df = pd.read_csv('nlp_result_modified.csv')

# 변환문 이해 정도와 관련된 항목들 간의 상관계수 계산
correlation = df[['Frequency of accessing economic news', 'Level of economic knowledge',
                  'Original text understanding level', 'Transformed text understanding level',
                  'Helpfulness of term explanation', 'Degree of content reflection']].corr()

# '변환문 이해 정도'와 다른 변수들 간의 상관 계수 확인
correlation_with_transformed = correlation['Transformed text understanding level']

# 상관 계수를 출력
print(correlation_with_transformed)

# 히트맵 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Variables Related to Transformed Text Understanding')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 로드
df = pd.read_csv('nlp_result_modified.csv')

# 변환문과 원문의 차이 계산
df['Difference'] = df['Transformed text understanding level'] - df['Original text understanding level']

# 차이 값에 대한 통계 출력
mean_difference = df['Difference'].mean()
max_difference = df['Difference'].max()
min_difference = df['Difference'].min()

print(f"차이 평균: {mean_difference:.2f}")
print(f"차이 최대값: {max_difference}")
print(f"차이 최소값: {min_difference}")

# 변환문이 원문보다 큰 경우
transformed_better = df[df['Difference'] > 0]
# 원문이 변환문보다 큰 경우
original_better = df[df['Difference'] < 0]

# 변환문이 원문보다 더 높은 응답자 수
transformed_better_count = len(transformed_better)
original_better_count = len(original_better)

print(f"변환문이 원문보다 큰 응답자 수: {transformed_better_count}명")
print(f"원문이 변환문보다 큰 응답자 수: {original_better_count}명")

# 히스토그램: 차이 분포 시각화
plt.figure(figsize=(8, 6))
plt.hist(df['Difference'], bins=10, color='skyblue', edgecolor='black')
plt.title('Difference between Transformed and Original Text Understanding Levels')
plt.xlabel('Difference (Transformed - Original)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 산점도: 원문과 변환문 간 차이를 시각화
plt.figure(figsize=(8, 6))

# 차이가 양수인 경우 (변환문 > 원문)
plt.scatter(transformed_better['Original text understanding level'], 
            transformed_better['Transformed text understanding level'], 
            color='green', label='Transformed > Original', alpha=0.7)

# 차이가 음수인 경우 (원문 > 변환문)
plt.scatter(original_better['Original text understanding level'], 
            original_better['Transformed text understanding level'], 
            color='red', label='Original > Transformed', alpha=0.5)

# 그래프 레이블 추가
plt.title('Comparison of Original and Transformed Text Understanding')
plt.xlabel('Original Text Understanding Level')
plt.ylabel('Transformed Text Understanding Level')

# 범례 추가
plt.legend()

# 그래프 표시
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# CSV 파일 로드 (nlp_result_with_diff.csv)
df = pd.read_csv('nlp_result_with_diff.csv')

# 상관계수를 계산할 변수들 리스트
columns_to_check = ['Frequency of accessing economic news', 'Level of economic knowledge',
                    'Original text understanding level', 'Transformed text understanding level',
                    'Helpfulness of term explanation', 'Degree of content reflection', 'diff']

# 해당 변수들 간의 상관계수 계산
correlation = df[columns_to_check].corr()

# 상관계수 출력
print("Correlation Matrix:")
print(correlation)

# 히트맵 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Variables (Including diff)')
plt.show()
```
